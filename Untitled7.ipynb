```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dataseti=pd.read_csv("/content/Ads_CTR_Optimisation.csv")
import math
N=100
d=10
ads_sel=[]
numbers_of_sel=[0]*d
sums_of_rewards=[0]*d
total_reward=0
for n in range(0,N):
  ad=0
  max_upper_bound=0
  for i in range(0,d):
    if(numbers_of_sel[i]>0):
      avg_rew=sums_of_rewards[i]/numbers_of_sel[i]
      delta_i=math.sqrt(3/2*math.log(n+1)/numbers_of_sel[i])
      upper_b=avg_rew+delta_i
    else:
      upper_b=float('inf')
    if(upper_b>max_upper_bound):
      max_upper_bound=upper_b
      ad=i
  ads_sel.append(ad)
  numbers_of_sel[ad]+=1
  reward=dataseti.values[n,ad]
  sums_of_rewards[ad]+=reward
  total_reward+=reward
plt.figure(figsize=(10,6))
plt.hist(ads_sel)
plt.title("Histogram of ads selections")
plt.xlabel("Ads")
plt.ylabel("Number of times each ad was selected")
plt.show()
```
